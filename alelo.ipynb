```python
#importa bibliotecas
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#carrega arquivis
heroesinfo=pd.read_csv("./heroes_information.csv") #info heros
shpower=pd.read_csv("./super_hero_powers.csv") #super heroes poweres
```

```python
heroesinfo.head(10)
shpower.head(10)
```

```python
heroesinfo.info()
```

```python
shpower.info()
```

```python
#recodifica vazios
heroesinfo = heroesinfo.replace(-99, np.nan)
heroesinfo = heroesinfo.rename(columns={'name': 'hero_names'})
```

```python
#fazer join das bases
sup_power_info = pd.merge(heroesinfo, shpower, on='hero_names', how='inner')
```

```python
sup_power_info.head(10)
```

```python
#arruma variaveis e trabalhar redução de dimenão pois são muitas colunas
shpower_cols = shpower.columns[1:]              # colunas dos poderes
sup_power_info[shpower_cols] = sup_power_info[shpower_cols].astype(int)  # True/False -> 1/0

num_cols = ['Height', 'Weight']
features = list(shpower_cols) + num_cols

X = sup_power_info[features].copy()
X[num_cols] = X[num_cols].fillna(X[num_cols].median())  # imputação
```

```python
#padronizando informações
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# fazendo a pca com 95$ de variancia
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("base original:", X.shape)
print("base após usar PCA:", X_pca.shape)
```

```python
# fazerdo clusters usando a base com redução de dimesão
ks = range(2, 11)
inertias = []
sil_scores = []

for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_pca)
    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(X_pca, labels))

# Exemplo: escolher o K que maximiza o silhouette
best_k = ks[int(np.argmax(sil_scores))]
print("Melhor K (pelo silhouette):", best_k)
```

```python
# kmeans final
final_km = KMeans(n_clusters=best_k, random_state=42, n_init=10)
sup_power_info['cluster'] = final_km.fit_predict(X_pca)
```

```python

#perfils já clasterizados
cluster_profiles = []

for c in range(best_k):
    subset = sup_power_info[sup_power_info['cluster'] == c]
    mean_powers = subset[shpower_cols].mean().sort_values(ascending=False)
    top_powers = mean_powers.head(10)
    median_height = subset['Height'].median()
    median_weight = subset['Weight'].median()
    size = len(subset)

    cluster_profiles.append({
        'cluster': c,
        'size': size,
        'median_height': median_height,
        'median_weight': median_weight,
        'top_powers': top_powers
    })

# Exemplo: olhar o perfil do primeiro cluster
for profile in cluster_profiles:
    print(f"\nCluster {profile['cluster']}")
    print("Tamanho:", profile['size'])
    print("Altura mediana:", profile['median_height'])
    print("Peso mediano:", profile['median_weight'])
    print("Principais poderes:")
    print(profile['top_powers'])
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python

#targets bons

# Filtrar apenas good e bad
sph2 = sup_power_info[sup_power_info['Alignment'].isin(['good', 'bad'])].copy()
sph2['target'] = sph2['Alignment'].map({'good': 1, 'bad': 0})

# Usar apenas variáveis binárias (poderes)
shpower_cols = list(shpower.columns[1:])   # todas as colunas de poderes
X = sph2[shpower_cols]                     # apenas 0/1 -> perfeito para BernoulliNB
y = sph2['target']

# Split com estratificação
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- Modelo ---

model = BernoulliNB()
model.fit(X_train, y_train)

# --- Avaliação ---

y_pred = model.predict(X_test)

print("Acurácia:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#randon forest regressos
# ---------------------------------------------
# 1. Imports
# ---------------------------------------------
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

# ---------------------------------------------
# 2. Pré-processamento
# ---------------------------------------------

df_reg = sup_power_info.copy()

# Manter pesos válidos
df_reg = df_reg[df_reg['Weight'].notna()]
df_reg = df_reg[df_reg['Weight'] > 0]    # remover pesos inválidos

# Seleção das features:
power_cols = list(shpower.columns[1:])
features = power_cols + ['Height']

X = df_reg[features].copy()
y = df_reg['Weight']

# ---- CORREÇÃO: tratar NaN ----

# Preencher NaN nos poderes com 0
X[power_cols] = X[power_cols].fillna(0)

# Preencher NaN na altura com mediana
X['Height'] = X['Height'].fillna(X['Height'].median())

# Garantir que tudo é numérico
X = X.astype(float)

# ---------------------------------------------
# 3. Split
# ---------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ---------------------------------------------
# 4. Modelo
# ---------------------------------------------
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

# ---------------------------------------------
# 5. Métricas
# ---------------------------------------------
y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE  (Erro Absoluto Médio):", round(mae,2))
print("RMSE (Raiz do Erro Quadrático Médio):", round(rmse,2))
print("R²   (Coeficiente de Determinação):", round(r2,3))

# ---------------------------------------------
# 6. Importância das variáveis
# ---------------------------------------------
importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
print("\nTop 10 variáveis mais importantes:")
print(importances.head(10))
```
